        Fourier Transform
    </h2>
</div>

***
The Fourier Transform is a mathematical technique used to break down a function into a combination of sine and cosine functions of various frequencies, thereby revealing the frequency components of the function or signal.

In image processing, the 2-dimensional Fourier Transform decomposes image data into its constituent frequencies, enabling manipulation of the image in the frequency domain - this can be used to remove high-frequency noise or enhance certain features. The high-frequency noise are identifiable spike points which are then masked out.

```python
def preview_fft(c_ft, c_clean, cmap):
    """Previews the image in fft dimension and color channel
    """
    ratio = c_ft.shape[0] / c_ft.shape[1]
    fig, axs = plt.subplots(2, 2, figsize=(15, 9*ratio))
    axs = axs.flatten()

    axs[0].imshow(np.log(abs(c_ft)), cmap='gray')
    axs[1].imshow(np.abs(np.fft.ifft2(c_ft)), cmap=cmap)
    axs[2].imshow(np.log(abs(c_clean)), cmap='gray')
    axs[3].imshow(np.abs(np.fft.ifft2(c_clean)), cmap=cmap)

    for ax in axs:
        ax.axis('off')
    cmap = 'grays' if cmap == 'gray' else cmap
    fig.tight_layout()
    toc.add_fig('Sample Fourier Transform + Suppression', width=100)
    
    
def suppress(image_fft, percentile=99.9, c_percent=0.05):
    """Suppress the patterns in the spatial domain
    """
    
    transformed_fft = image_fft.copy()
    thres = np.percentile(np.log(np.abs(transformed_fft)).reshape(-1),
                          percentile)
    
    mask = np.ones((image_fft.shape[0], image_fft.shape[1]))
    center_x, center_y = image_fft.shape[0] / 2, image_fft.shape[1] / 2
    mask[int(center_x*(1-c_percent)):int(center_x*(1+c_percent)),
         int(center_y*(1-c_percent)):int(center_y*(1+c_percent))] = False

    transformed_fft[np.logical_and(np.log(abs(transformed_fft)) >= thres,
                                   mask)] = 1
    
    return transformed_fft
```

```python
image = imread('../moon_image.png')
fig, ax = plt.subplots(figsize=(15,8))
imshow(image)
ax.axis('off')
toc.add_fig('Sample Image (Moon)', width=100)
```

```python
img_ft = np.fft.fftshift(np.fft.fft2(image))
img_clean = suppress(img_ft, percentile=99.8)
preview_fft(img_ft, img_clean, cmap='gray')
```

***

The purpose of white balancing is to remove strong color casts caused by different light sources, such as tungsten (indoor lighting), fluorescent, or natural daylight.

        Ground Truth
    </h3>

***

This algorithm involves using a white reference patch or card to establish the correct color balance in an image or video. We feed the algorithm with a reliable reference for true white, allowing it to adjust the color temperature and remove any color casts present in the lighting conditions.

```python
def ground_truth_set(image, patch, name):
    """White balance image using Ground-truth algorithm

    Parameters
    ----------
    image : numpy array
        Image to white balance
    patch : numpy array
        Patch of "true" white

    Returns
    -------
    None
    """
    x1, y1 = patch[0]
    x2 = x1 + patch[1]
    y2 = y1 + patch[2]
    
    image_gt_mean = (image*1.0 /
                     image[x1:x2, y1:y2].mean(axis=(0, 1))).clip(0, 1)
    image_gt_max = (image*1.0 /
                    image[x1:x2, y1:y2].max(axis=(0, 1))).clip(0, 1)
        
    fig, ax = plt.subplots(1, 3, figsize=(15,6))
    ax[0].imshow(image[x1:x2, y1:y2])
    ax[0].axis('off')
    ax[0].set_title(f'Ground Truth: {name}')
    
    ax[1].imshow(image_gt_mean)
    ax[1].add_patch(Rectangle((patch[0]), patch[2], patch[1],
                              edgecolor='r',
                              facecolor='none'))
    ax[1].set_title('Ground Truth (Mode: Mean)')
    ax[1].axis('off')
    
    ax[2].imshow(image_gt_max)
    ax[2].add_patch(Rectangle((patch[0]), patch[2], patch[1],
                              edgecolor='r',
                              facecolor='none'))
    ax[2].set_title('Ground Truth (Mode: Max)')
    ax[2].axis('off')
    toc.add_fig('Set of Sample White Balancing using Ground Truth', width=100)
```

        White Patch Algorithm

***
The white patch algorithm is based on the idea that in a color image, white is represented by the brightest possible color values. In a 24-bit image, these values are (255, 255, 255) for the red, green, and blue channels. So, to achieve a proper white balance, the algorithm adjusts each channel independently by making sure it reaches its maximum brightness.

```python
def white_patch_set(image, percentile=100):
    """White balance image using White patch algorithm
    
    Parameters
    ----------
    image : numpy array
        Image to white balance
    percentile : integer, optional
        Percentile value to consider as channel maximum
        
    Returns
    -------
    None
    """
    fig, axs = plt.subplots(1, 3, figsize=(15,6))
    for i, ax in enumerate(axs[::-1]):
        image_wp = img_as_ubyte((image*1.0 /
                                 np.percentile(image,
                                                percentile-2.5*(i+1),
                                                axis=(0, 1)))
                                 .clip(0, 1))
        ax.imshow(image_wp)
        ax.axis('off')
        ax.set_title(f'White Patch (Percentile: {percentile-2.5*(i+1):.2f})')
        
    toc.add_fig('Set of Sample White Balancing using White Patch', width=100)
```

***
The Gray World algorithm is a color correction technique used in image processing to achieve a more balanced and natural color representation in an image. It assumes that, on average, the colors in a scene should appear gray, and it adjusts the image accordingly.

```python
def gray_world(image):
    """White balance image using Gray-world algorithm

    Parameters
    ----------
    image : numpy array
        Image to white balance

    Returns
    -------
    None
    """
    image_gw = ((image * (image.mean() / image.mean(axis=(0, 1))))
             .clip(0, 255).astype(int))
    plt.imshow(image_gw[:,:,:])
    plt.axis('off')
    toc.add_fig('Gray World Algorithm')
```

***
Histogram manipulation refers to the process of modifying the distribution of pixel intensities in an image by adjusting its histogram. The histogram represents the frequency of occurrence of different intensity values in an image. By manipulating the histogram, it is possible to enhance image contrast, adjust brightness levels, improve dynamic range, and perform other modifications to achieve desired visual effects or enhance image quality. Common histogram manipulation techniques include histogram equalization, contrast stretching, and gamma correction. These methods can help optimize the overall appearance and enhance the details and features within an image.

```python
image3 = imread('../dark_image.png')
fig, ax = plt.subplots(figsize=(15,8))
imshow(image3)
ax.axis('off')
toc.add_fig('Original Image for Histogram Manipulation')
```

        Histogram Eqaulization
    </h3>

***
Redistributes the pixel intensities in the image's histogram to create a more balanced distribution across the available intensity range. By doing so, histogram equalization can enhance both dark and bright regions, bringing out more details and improving the overall image quality.

```python
def rescale_image(image, target_freq, target_bins, cmap):
    """Rescales a given image channel to the given target parameters
    """
    fig, axs = plt.subplots(1, 3, figsize=(15, 3.3))

    freq, bins = cumulative_distribution(image)

    new_vals = np.interp(freq, target_freq, target_bins)
    new_image = img_as_ubyte(new_vals[image].astype(int))
    imshow(new_image);

    axs[0].imshow(image, cmap=cmap)
    axs[0].set_title(f'Original Channel ({cmap[:-1]})')
    axs[0].axis('off')

    axs[1].step(bins, freq, c='y', label='actual cdf')
    axs[1].plot(target_bins, target_freq, c='k', label='target cdf')
    axs[1].legend()
    axs[1].set_xlim(0, 255)
    axs[1].set_ylim(0, 1)
    axs[1].set_xlabel('intensity values')
    axs[1].set_ylabel('cumulative fraction of pixels')

    axs[2].imshow(new_image, cmap=cmap)
    axs[2].set_title(f'Manipulated Channel ({cmap[:-1]})')
    axs[2].axis('off')
    
    return new_image
```

```python
cmaps = ['Reds', 'Greens', 'Blues']

target_bins = np.arange(255)
target_freq = np.linspace(0, 1, len(target_bins))

linear_image = image3.copy()
for i, cmap in enumerate(cmaps):
    linear_image[:,:,i] = rescale_image(image3[:,:,i],
                                        target_freq,
                                        target_bins,
                                        cmap)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax[0].imshow(image3)
ax[0].set_title('Before')
ax[0].axis('off')
ax[1].imshow(linear_image)
ax[1].set_title('After')
ax[0].axis('off')
toc.add_fig('Histogram Equalization', width=100)
```

        Contrast Stretching
    </h3>
</div>

***
By rescaling the intensity values, the darkest pixels become even darker, and the brightest pixels become even brighter, resulting in a more pronounced distinction between different regions or objects. Contrast stretching is often employed to enhance images with low contrast or to bring out finer details in specific areas of interest.

```python
def contrast_stretch(image, low=5, high=95):
    """Perform constrast stretching on the image"""
    fig, axs = plt.subplots(1, 2, figsize=(15, 4))

    img_contrast = (
        rescale_intensity(image,
                          in_range=tuple(np.percentile(image,
                                                       (low, high))))
    )

    axs[0].imshow(image, cmap=cmap)
    axs[0].set_title(f'Original')
    axs[0].axis('off')
    
    axs[1].imshow(img_contrast, cmap=cmap)
    axs[1].set_title(f'Contrast Stretched')
    axs[1].axis('off')
    
    return img_contrast
```

```python
contrast_image = image3.copy()
for i, cmap in enumerate(cmaps):
    contrast_image[:,:,i] = contrast_stretch(image3[:,:,i])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax[0].imshow(image3)
ax[0].set_title('Before')
ax[0].axis('off')
ax[1].imshow(contrast_image)
ax[1].set_title('After')
ax[1].axis('off')
toc.add_fig('Constrast Stretching', width=100)
```
